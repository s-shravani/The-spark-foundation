Comment: Here country is only of one unique values i.e. United States. Thus we can drop this column

```python
# Deleting the unwanted columns
c = ['Postal Code', 'Country']   #postal code is not required for analysis
retail = data.drop(columns=c,axis=1)
retail.head()
```

## Data Analysis and Visualization

```python
# Lineplot between Profit and Discount
plt.figure(figsize=(8,6))
sns.lineplot(x='Discount',y='Profit',data=retail)
plt.show()
```

Comment : Here we can see the trend of profit with respect to discount value.

```python
# Calculating the statewise profit
profit_sum = retail.groupby("State")["Profit"].sum()
profit_sum
```

```python
# State wise Profit of United States of America
plt.figure(figsize=(18,12))
sns.barplot(x=profit_sum.index, y=profit_sum.values)
plt.ylabel("Profit")
plt.xticks(rotation=90)
plt.show()
```

Inference:

Highest profit is shown by 'California' and lowest profit is shown by 'Texas'

Top 4 profit giving states: 'California','New York','Washington'and'Michigan'

Top 4 loss giving states:'Illinois','Pennsylvania','Ohio'and'Texas

```python
# Calculating the citywise profit
city_profit = retail.groupby('City')['Profit'].sum().sort_values(ascending = False)
city_profit
```

```python
# City wise Profit of United States of America
city_profit[:5].plot(kind='bar')
```

```python
city_profit[526:].plot(kind='bar')
```

```python
# State wise Sales of USA
state_sales = retail.groupby('State')['Sales'].sum()
state_sales
```

```python
# State wise Sales of United States of America
plt.figure(figsize=(18,12))
sns.barplot(x=state_sales.index, y=state_sales.values)
plt.ylabel("Sales")
plt.xticks(rotation=90)
plt.show()
```

Inference:

Highest sales is shown by 'California'

Top 4 States with highest Sales : 'California','New York','Texas'and 'Washington'

Top 4 states with lowest Sales:'North Dakota','Maine','West Virginia'and'South Dakota'

Thus, Texus has high sales but low profit

```python
# Comparing Sales and Profit State  wise

profit_sales = retail.groupby("State")["Profit", "Sales"].sum()
profit_sales.plot.bar(figsize=(10,6))
plt.show()
```

Inference:

Texas, Pennsylvania, and Ohio have high Sales but negative Profit i.e. loss

```python
# State wise Avg_discount_percent of USA
Avg_discount_percent = retail.groupby("State")["Discount"].mean()*100
Avg_discount_percent
```

```python
# State wise Avg_discount_percent of United States of America
plt.figure(figsize=(15,10))
sns.barplot(x=Avg_discount_percent.index, y=Avg_discount_percent.values)
plt.ylabel("Avg_discount_percent")
plt.xticks(rotation=90)
plt.show()
```

Inference: Illinois has the highest Avg_discount_percent among the states of USA

```python
# Distrubution of Profits of Segment across diffrent regions
plt.figure(figsize = (10,6))
sns.barplot(data = retail , x = "Region",y = "Profit" ,hue = "Segment")
plt.show()
```

Inference: Highest profits for each region are as follows:

South= Consumer Segment

West= Corporate Segment

Central= Home Office Segment

East= Home Office Segment

Lowest profits for each region are as follows:

South= Home Office Segment

West= Home Office Segment

Central= Consumer Segment

East= Corporate Segment

```python
# Segment Vs Category
x =retail['Segment']
y = retail['Category']
 
# setting the dimensions of the plot
fig, ax = plt.subplots(figsize=(12,8))
 
# drawing the plot
sns.lineplot(x, y, ax=ax)
plt.show()
```

```python
# Analyzing Category wise sales, Discount, Profit:


category_data = retail.groupby(['Category'])[['Sales', 'Discount', 'Profit']].mean()
category_data
```

```python
# Pie Charts of Category wise sales, Discount, Profit
category_data.plot.pie(subplots=True, figsize=(20, 18),autopct='%0.1f%%', labels = category_data.index)
plt.show()
```

Inference:

1)Technology Category has the maximum Sales and Profit.

2)Office Supplies category has minimum sales and Furniture category has lowest Profit.

```python
# Analyzing Ship Mode wise sales, Discount, Profit

shipmode_data = retail.groupby(['Ship Mode'])[['Sales', 'Discount', 'Profit']].mean()
shipmode_data.plot.pie(subplots=True,figsize=(18, 20), autopct='%0.1f%%', labels = shipmode_data.index)
plt.show()
```

Inference:

Same Day Ship Mode has the highest sales.

First Class Ship mode has the highest Profit and Discount

Standard Class Ship mode has lowest sales.

Standard Class Ship mode has lowest profit.

```python
# Checking the profits of 'Sub- Category'
pivot=pd.pivot_table(retail,index='Sub-Category',values='Profit')
pivot.plot(kind='bar')
plt.show()
```

```python
# Checking the Sales of 'Sub- Category'
pivot1=pd.pivot_table(retail,index='Sub-Category',values='Sales')
pivot1.plot(kind='bar')
plt.show()
```

Inference: Among 'Sub- Category' items, Copiers and machines have the highest Sales in United States.Here we see that copiers had highest sale and Machines had second highest sales.In above two graph we see that 'Machines' had second highest sale but due to large discount it has less profit.

Low Sales 'Sub- Category' items: Fasteners, Labels, Paper, Art

```python
# Drawing Heatmap to identify correlation between numerical columns:
plt.figure(figsize=(10,8))
correlation=retail.corr()
sns.heatmap(correlation,annot=True,cmap="RdYlGn")
plt.show()
```

Inference:

From the heatmap, we can see that there is positive correlation of 0.48 between Sales and Profit. Thus if Sales increases, Profit increases.

From the heatmap, we can see that there is negative correlation of -0.22 between Profit and Discount. Thus if Discount increases, most of the times Profit goes down.
